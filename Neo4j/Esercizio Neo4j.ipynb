```python
from neo4j import GraphDatabase, basic_auth
import pandas
```

```python
driver = GraphDatabase.driver(
  "bolt://3.227.231.122:7687",
  auth=basic_auth("neo4j", "diagonals-shots-incentive"))
```

```python

cypher_query = '''
call
db.schema.visualization()
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
result
```

```python
#1. i nomi delle squadre che hanno partecipano al torneo nel 2015
cypher_query = '''
MATCH (t:Team)-[:PARTICIPATED_IN]-(a:Tournament)
WHERE a.year = 2015
RETURN t
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
pandas.DataFrame(result)
```

```python
#2. i nomi delle squadre che hanno partecipato al torneo nel 2015 (farlo in un altro modo)
cypher_query = '''
MATCH (t:Team)-[:PARTICIPATED_IN]-(:Tournament {year:2015}) RETURN t
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
pandas.DataFrame(result)
```

```python
#3. il numero di partecipazioni di ogni nazione
cypher_query = '''
MATCH (t:Team)-[:REPRESENTS]-(p:Person) RETURN t,count(p)
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
pandas.DataFrame(result)
```

```python
#4. gli allenatori dell'Italia
cypher_query = '''
MATCH (p:Person)-[:COACH_FOR]-(s:Squad)
WHERE s.id STARTS WITH "Italy"
RETURN p.name
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
pandas.DataFrame(result)
```

```python
#5. le giocatrici dell'Italia che hanno segnato e in quale partita hanno segnato (data)
cypher_query = '''
MATCH (p:Person)-[:SCORED_GOAL]-(m:Match)
WHERE s.id STARTS WITH "Italy"
RETURN p.name
'''

with driver.session() as graphDB_Session:
    result = graphDB_Session.run(cypher_query).data()
pandas.DataFrame(result)
```
